```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randn(1000)

#Making the bins
width = .01
histmin = 0
histmax = 1.0

bins = np.arange(histmin, histmax, width)
```

Graph the Histogram

```python
plt.hist(x, bins = bins, edgecolor='black')
plt.xlabel('x')
plt.ylabel('N per bin')

#save pdf
plt.savefig("Histogram2.pdf", bbox_inches="tight")
```
