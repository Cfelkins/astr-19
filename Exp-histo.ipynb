```python
import numpy as np
import matplotlib.pyplot as plt
```

Make a list of 1000 random values using an exponential distribution

```python
x = np.random.exponential(1, 1000)
```

create histogram for x with 100 bins

```python
#define bins
width = 1
histmin = 0
histmax = 100

bins = np.arange(histmin, histmax, width)
```

Plot histogram

```python
plt.hist(x, bins=bins, edgecolor='black')
plt.xlabel('x')
plt.ylabel('N in box')

#Save pdf
plt.savefig('Exp-histo.pdf', bbox_inches="tight")
```
