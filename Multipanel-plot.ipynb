```python
import numpy as np
import matplotlib.pyplot as plt
```

make an array from 0 to 1 in 100 steps 

```python
x = np.linspace(0, 1, 100)
```

define functions returning sin and cos

```python
s = np.sin(x)
c = np.cos(x)
```

Make multipanel plot and label axes

```python
f, axarr = plt.subplots(1, 2)

axarr[0].plot(x, s)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')

axarr[1].plot(x, c)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')

#spacing graphs apart
f.subplots_adjust(wspace=0.5)
```

Save file as pdf

```python
plt.savefig("Multipanel-subplots.pdf", bbox_inches='tight', facecolor='white')
```
